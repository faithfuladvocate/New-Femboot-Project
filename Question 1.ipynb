```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(10)
x=np.random.rand(20)
```

```python
y=[]
def value(n):
    y.append(n**3 + np.sin(n))
    return y
value(x)
```

```python
y1 = np.round(y) 
```

```python
y1 = y1.reshape(20,)
```

```python
plt.plot(x,y1)
```

```python
plt.bar(y1, x)
```
